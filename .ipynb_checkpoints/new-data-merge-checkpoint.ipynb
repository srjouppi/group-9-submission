# Geocoding Sources with Census Tract-Level Data

```python
# Source List
source_list = pd.read_excel("CMS_Subject_Sources_Sheet2.xlsx")
source_list.columns = source_list.columns.str.lower().str.replace(" ","_")
# Adding a Long address to merge with geocoded addresses
source_list['long_address'] = source_list.address + ", " + source_list.city

# Geocoded Sources with Census Tract Geoids (using QGIS join by location)
facilities_geocoded_census = pd.read_csv("facilities-census-tract-2.csv",dtype={"GEOID":"str"})
facilities_geocoded_census.columns = facilities_geocoded_census.columns.str.lower().str.replace(" ","_")
# Shortening the address to
facilities_geocoded_census['simple_address'] = facilities_geocoded_census.address.str.split(", MI").str[0]

census_data_race = pd.read_csv("DECENNIALPL2020.P1_data_with_overlays_2022-05-02T140811.csv",header=1)
census_data_race.columns = census_data_race.columns.str.lower().str.replace(" ","_").str.replace("!!","").str.replace(":","_")
census_data_race['geoid'] = census_data_race.id.str[-11:]

census_data_income = pd.read_csv("ACSST5Y2020.S1901_data_with_overlays_2022-05-15T023414.csv",header=1,na_values="-")
census_data_income.columns = census_data_income.columns.str.lower().str.replace(" ","_").str.replace("!!","").str.replace(":","_")
census_data_income['geoid'] = census_data_income.id.str[-11:]
```

```python
sources_geocoded = source_list.merge(facilities_geocoded_census,how="left",left_on="long_address",right_on="simple_address")

missing_addresses = sources_geocoded.query("address_y.isnull()").address_x.to_list()

sources_geocoded = sources_geocoded.query("address_y.notnull()")

sources_geocoded_2 = source_list.merge(facilities_geocoded_census,how="left",left_on="address",right_on="simple_address").query('address_x.isin(@missing_addresses)')
sources_geocoded_all = pd.concat([sources_geocoded,sources_geocoded_2],ignore_index=True)
```

```python
facilities_geocoded_census.query("address.str.contains('selfridge',case=False)")
```

```python
sources_simple = sources_geocoded_all[['srn','geoid','name']]

race_simple = census_data_race[['geoid','_total_','_total_population_of_one_race_','_total_population_of_one_race_white_alone','_total_population_of_one_race_black_or_african_american_alone']]
race_simple = race_simple.rename({'_total_population_of_one_race_':'pop_total_onerace','_total_population_of_one_race_white_alone':'white_total','_total_population_of_one_race_black_or_african_american_alone':'black_total'},axis=1)
race_simple['pct_white'] = (race_simple.white_total / race_simple.pop_total_onerace).round(4)*100
race_simple['pct_black'] = (race_simple.black_total / race_simple.pop_total_onerace).round(4)*100

income_simple = census_data_income[['geoid','estimatehouseholdsmedian_income_(dollars)']]
income_simple = income_simple.rename({'estimatehouseholdsmedian_income_(dollars)':'median_income_dollars'},axis=1)
income_simple.median_income_dollars = income_simple.median_income_dollars.astype("float")
```

```python
sources_demo = sources_simple.merge(race_simple,how="left",left_on="geoid",right_on="geoid")
sources_demo = sources_demo.merge(income_simple,how="left",left_on="geoid",right_on="geoid")
sources_demo
```
